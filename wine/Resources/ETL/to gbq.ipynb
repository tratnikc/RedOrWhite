```python
#  ETL to google big query

from google.oauth2 import service_account
import pandas as pd
import pandas_gbq
import json

## credentials
projectID = 'project-wine-308217'
datasetID = 'wine_dataset'
bigquery_uri = f'bigquery://{projectID}/{datasetID}'

credentials = service_account.Credentials.from_service_account_file('project-wine.json')

# step1
# open csv from_service_account_file
red= pd.read_csv('red.csv', encoding='latin-1')
white= pd.read_csv('white.csv', encoding='latin-1')



# step2
# load csv file to dataframe

```

```python
red1=red[red.columns[0:13]]
white1=white[white.columns[0:13]]
```

```python
red2=red1.rename(columns={'title': 'wine_name', 'link' : 'link','Img' :'img' , 'Bottle Size': 'bottle_size_ml', 'price': 'price', 'Alcohol/Vol:' : 'alcohol_content', 'Made In:' : 'country', 'By:': 'winery', 'Sugar Content:': 'sugar_content_gL',
                         'Sweetness Descriptor:': 'sweetness_descr', 'Style:':'style', 'Varietal:':'varietal', 'Description': 'wine_desc'})

white2=white1.rename(columns={'title': 'wine_name', 'link' : 'link','Img' :'img' , 'Bottle Size': 'bottle_size_ml', 'price': 'price', 'Alcohol/Vol:' : 'alcohol_content', 'Made In:' : 'country', 'By:': 'winery', 'Sugar Content:': 'sugar_content_gL',
                         'Sweetness Descriptor:': 'sweetness_descr', 'Style:':'style', 'Varietal:':'varietal', 'Description': 'wine_desc'})
```

```python
white2
```

```python
red2['price'] =red2['price'].str.replace('$','').replace('$', '')
red2['alcohol_content'] =red2['alcohol_content'].str.replace('%','')
red2['sugar_content_gL'] =red2['sugar_content_gL'].str.replace('g/L','')

white2['price'] =white2['price'].str.replace('$','').replace('$', '')
white2['alcohol_content'] =white2['alcohol_content'].str.replace('%','')
white2['sugar_content_gL'] =white2['sugar_content_gL'].str.replace('g/L','')





```

```python
red2['price'] =red2['price'].str.replace(',', '')
white2['price'] =white2['price'].str.replace(',', '')
```

```python
red2['price'] = red2['price'].str[:5]
white2['price'] = white2['price'].str[:5]

```

```python
red2[["price", "alcohol_content", 'sugar_content_gL']] =red2[["price", "alcohol_content", 'sugar_content_gL']].apply(pd.to_numeric)
white2[["price", "alcohol_content", 'sugar_content_gL']] =white2[["price", "alcohol_content", 'sugar_content_gL']].apply(pd.to_numeric)
```

```python
red2['category']='red'
```

```python
red3=red2[['wine_name', 
                'link',
                'img',
                'bottle_size_ml',
                'alcohol_content',
                'country', 
                'winery', 
                'sugar_content_gL', 
                'sweetness_descr', 
                'style', 
                'price',
           'category', 
           'wine_desc'
               ]]
```
