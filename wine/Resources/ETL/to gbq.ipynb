```python
#  ETL to google big query

from google.oauth2 import service_account
import pandas as pd
import pandas_gbq
import json

## credentials
projectID = 'project-wine-308217'
datasetID = 'wine_dataset'
bigquery_uri = f'bigquery://{projectID}/{datasetID}'

credentials = service_account.Credentials.from_service_account_file('project-wine.json')

# step1
# open csv from_service_account_file
red= pd.read_csv('red.csv', encoding='latin-1')
white= pd.read_csv('white.csv', encoding='latin-1')



# step2
# load csv file to dataframe

```

```python
red1=red[red.columns[0:13]]
white1=white[white.columns[0:13]]
```

```python
red1
```

```python
red2=red1.rename(columns={'title': 'Title', 'link' : 'Link', 'price': 'Price ($)', 'Alcohol/Vol:' : 'Alcohol/Vol (%)', 'Made In:' : 'Made In', 'By:': 'By', 'Sugar Content:': 'Sugar Content (g/L)',
                         'Sweetness Descriptor:': 'Sweetness Descriptor', 'Style:':'Style', 'Varietal:':'Varietal'})

white2=white1.rename(columns={'title': 'Title', 'link' : 'Link', 'price': 'Price ($)', 'Alcohol/Vol:' : 'Alcohol/Vol (%)', 'Made In:' : 'Made In', 'By:': 'By', 'Sugar Content:': 'Sugar Content (g/L)',
                         'Sweetness Descriptor:': 'Sweetness Descriptor', 'Style:':'Style', 'Varietal:':'Varietal'})
```

```python
red2
```

```python
red2['Price ($)'] =red2['Price ($)'].str.replace('$','').replace('$', '')
red2['Alcohol/Vol (%)'] =red2['Alcohol/Vol (%)'].str.replace('%','')
red2['Sugar Content (g/L)'] =red2['Sugar Content (g/L)'].str.replace('g/L','')

white2['Price ($)'] =red2['Price ($)'].str.replace('$','').replace('$', '')
white2['Alcohol/Vol (%)'] =red2['Alcohol/Vol (%)'].str.replace('%','')
white2['Sugar Content (g/L)'] =red2['Sugar Content (g/L)'].str.replace('g/L','')





```

```python
red2.head()
```

```python
red2['Price ($)'] =red2['Price ($)'].str.replace(',', '')
white2['Price ($)'] =white2['Price ($)'].str.replace(',', '')
```

```python
red2['Price ($)'] = red2['Price ($)'].str[:5]
white2['Price ($)'] = white2['Price ($)'].str[:5]

```

```python
red2[["Price ($)", "Alcohol/Vol (%)", 'Sugar Content (g/L)']] =red2[["Price ($)", "Alcohol/Vol (%)", 'Sugar Content (g/L)']].apply(pd.to_numeric)
white2[["Price ($)", "Alcohol/Vol (%)", 'Sugar Content (g/L)']] =white2[["Price ($)", "Alcohol/Vol (%)", 'Sugar Content (g/L)']].apply(pd.to_numeric)
```
